# Plotting the Graphs

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(housing_data['Sale Price'])
```

```python
#ADDing x and y axis label , and title to the graph
plt.plot(housing_data['Sale Price'])
plt.xlabel("Record Number")
plt.ylabel("Sale Price")
plt.title("My first graph")
plt.show()
```

```python
plt.plot(housing_data['Sale Price'],marker="o",color="green",markerfacecolor="Blue",markersize="5",linewidth="5",linestyle="dashed")
plt.xlabel("Record Number")
plt.ylabel("Sale Price")
plt.title("My first graph")
plt.show()
```

```python
housing_data.groupby('Condition of the House')['ID'].count()
```

```python
#Plotting Pie chart out of the data received
values=(30,1701,14031,5679,172)
labels=('Bad','Excellent','Fair','Good','Okay')
plt.pie(values,labels=labels)
```

```python
#bar graph for above data
plt.bar(labels,values,color="Blue")
plt.xlabel("Conditions")
plt.ylabel("Numbers")
plt.title("GRaph for condtions")
plt.show()
```

```python
plt.scatter(x=housing_data['Flat Area (in Sqft)'],y=housing_data['Sale Price'],color='red')
plt.xlabel=("Area")
plt.ylabel=("Selling Price")
plt.title("Selling Price VS Area")
plt.show()
```
